```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
meta=pd.read_csv('/content/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv')
meta
```

```python
region_counts = meta['Region'].value_counts()
```

```python
plt.bar(region_counts.index, region_counts.values)
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.title('Distribution of Countries by Region')
plt.xticks(rotation=45,ha='right')
plt.show()
```

```python
income_counts = meta['IncomeGroup'].value_counts()

plt.figure(figsize=(5, 5))
income_counts.plot(kind='bar', width=0.5)
plt.title('Distribution of Countries by Income Group')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```
